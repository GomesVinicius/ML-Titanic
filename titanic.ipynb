```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

import numpy as np
```

```python
df = pd.read_csv('data/titanic.csv')
X = df.drop(['Survived', 'Name'], axis=1)
y = df.Survived
```

```python
onehot = OneHotEncoder(sparse_output=False, drop='first')

X_is_objects = onehot.fit_transform(X.select_dtypes(include=['object']))
X_not_is_objects = X.select_dtypes(exclude=['object'])

X_all = np.append(X_not_is_objects, X_is_objects, axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_all, y, train_size=0.8, random_state=42)

knn = KNeighborsClassifier(n_neighbors=8)
knn.fit(X_train, y_train)

accuracy_score(y_test, knn.predict(X_test))
```

Resultado de 68.53% sem normalização dos dados

Subindo para 70.22% ao tratar a coluna 'Sex' como binário

Subindo para 75.84% removendo a coluna 'Fare'

```python
normalizer = MinMaxScaler()
X_norm = pd.DataFrame(
    columns=['Pclass', 'Age', 'Siblings/Spouses Aboard', 'Parents/Children Aboard', 'Fare', 'Sex'],
    data=normalizer.fit_transform(X_all))
X_norm
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_norm, y, train_size=0.8, random_state=42)

knn.fit(X_train, y_train)
accuracy_score(y_test, knn.predict(X_test))
```

Resultado de 70.78% COM normalização dos dados

Subindo para 76.96% ao tratar a coluna 'Sex' como binário

Subindo para 77.52% removendo a coluna 'Fare'

Porém, no fim, tratando com um bom número de vizinhos (4) a acuracidade se mantém ligeiramente maior ao tratar com a coluna 'Fare'

```python
max_value_accuracy_score = []

for x in range(8):
    knn = KNeighborsClassifier(n_neighbors=x+1)
    knn.fit(X_train, y_train)
    score = accuracy_score(y_test, knn.predict(X_test))
    max_value_accuracy_score.append(score)
    print(x+1, score)

print(f'{max(max_value_accuracy_score) * 100:.2f} % com 4 vizinhos')
```
